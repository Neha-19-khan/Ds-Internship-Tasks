## Titanic Dataset - EDA and Visualization



**Step 1: Import Libraries**


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Step 2: Load the dataset**

```python
df = pd.read_csv(r"C:\Users\Shahdil\Downloads\Titanic\tested.csv")
df
```

**Step 3: Basic Exploration**

```python
print(df.head())
print(df.info())
print(df.describe())
```

**Step 4: Data Cleaning**

**1.Handle missing values**

```python
print("\nMissing values before:")
print(df.isnull().sum())
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
```

**2.Remove duplicates**

```python
df.drop(columns=['Cabin'], inplace=True, errors='ignore')
df.drop_duplicates(inplace=True)
df
```


**Step 5: Outlier Detection**

```python
sns.boxplot(x=df['Fare'])
plt.title('Fare Boxplot')
plt.show()
```


**Step 6: Visualizations**

 **1. Bar chart for 'Sex'**

```python
df['Sex'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Gender Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

**2. Histogram for 'Age'**

```python
df['Age'].hist(bins=30, color='orange')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

**3. Correlation Heatmap**

```python
numeric_df = df.select_dtypes(include=['number']) 
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# Insights 


- **Gender Distribution:** Majority of the passengers on board were male.
- **Age Distribution:** The age distribution is right-skewed, indicating more younger passengers compared to older ones.
- **Missing Values:** Age and Embarked columns had missing values, which were handled by imputing the mean for Age and mode for Embarked. The Cabin column was dropped due to excessive missing data.
- **Duplicates:** Duplicate records were removed to ensure data quality.
- **Fare Distribution:** Fare values show presence of outliers, as seen in the boxplot.
- **Correlation Analysis:** 
  - There is a strong negative correlation between `Pclass` and `Fare`, indicating that higher-class passengers paid significantly higher fares.
  - Moderate correlation exists between `Fare` and `Survived`, suggesting that passengers who paid higher fares had better chances of survival.
- **Additional Notes:**
  - Most passengers embarked from port 'S' (Southampton), which was the most common point of boarding.
  - The dataset reflects a typical passenger distribution of the Titanic tragedy, useful for predictive modeling.
